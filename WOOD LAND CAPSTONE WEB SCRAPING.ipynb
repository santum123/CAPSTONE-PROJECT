```python
last_height = driver.execute_script("return document.body.scrollHeight")
while True:
    # Scroll down to the bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    
    # Get the new height of the page
    new_height = driver.execute_script("return document.body.scrollHeight")

    # Break the loop if you have reached the bottom of the page
    if new_height == last_height:
        break

    # Update the last height
    last_height = new_height
```

```python
name=[]
price = []
colors = []
color_name=[]
total_sizes=[]
count_sizes = []
features=[]
comfort = []
product_code=[]

for product in product_urls:
    driver.get(product)
    time.sleep(10)
    
    #Extracting Product_code=============================
    code = urlparse(product).path
    product_code.append(code.split('/')[-1])
    
    # Extracting Name=================================================================================
    try:
        title = driver.find_element(By.XPATH,'//h2[@class="productprice_productname__7Lzg5"]')
        name.append(title.text)
    except NoSuchElementException:
        name.append('NA')
    
    #Extracting Price================================================================================
    try:
        Price_tag = driver.find_element(By.XPATH,'//div[@class="productprice_productprice__IuYyr"]')
        
        try:
            discount = Price_tag.find_element(By.XPATH,'.//span[@class="productprice_offerprice__RLPOw"]')
            price.append(discount.text.split('₹')[1])
        except NoSuchElementException:
            try:
                original =Price_tag.find_element(By.XPATH,'.//span[@class="productprice_mrpprice__7zMlK"]')
                price.append(original.text.split('₹')[1])
            except NoSuchElementException:
                price.append('NA')
                
    except (StaleElementReferenceException, NoSuchElementException):
        price.append('NA')
        
    #Extracting colors ===============================================================
    try:
        col_tags = driver.find_element(By.XPATH,'//div[@class="css-17outbb"][1]')
        tags = len(col_tags.find_elements(By.XPATH,'.//div'))
        colors.append(tags)
    except NoSuchElementException:
        colors.append('NA')
        
    #Extracting color names================================================================
    col_nam = driver.find_element(By.XPATH,'//h5[@class="colorpicker_colorname__sRdk8"]')
    color_name.append(col_nam.text)
    
    
    # Extracting sizes and count of sizes===========================================================
    
    try:
        size_containers = driver.find_element(By.XPATH, '//div[@class="MuiBox-root css-acwcvw"]/div')
        sizes = size_containers.find_elements(By.XPATH, '//div[@class="sizepicker_sizecontainer__xrprC"]')

        current_sizes = []
        for size in sizes:
            size_buttons = size.find_elements(By.XPATH, './/button[not(@disabled)]')
            if size_buttons:
                sizes_text = [button.text for button in size_buttons]
                current_sizes.extend(sizes_text)
        total_sizes.append(current_sizes)
        count_sizes.append(len(current_sizes))

    except NoSuchElementException:
        total_sizes.append('NA')
        count_sizes.append('NA')
        

    #Scroll========================================================
    scroll= driver.find_element(By.XPATH,'//div[@class="MuiBox-root css-15g2oxy"][2]')
    driver.execute_script("arguments[0].scrollIntoView(true);", scroll)
    time.sleep(2)
    #Expanding the product details======================================================
    expand = driver.find_element(By.XPATH,'//div[@id="panel1bh-header"]')
    expand.click()
    time.sleep(5)
    
    # Extracting Features================================================================
    try:
        des = driver.find_element(By.XPATH,'//div[@class="MuiBox-root css-t7e7gw"]/div/ul')
        features.append(des.text)
    except NoSuchElementException:
        features.append('NA')
        
    #Extracting comfort===================================================================
    try:
        comfi_elements = driver.find_elements(By.XPATH, '//div[@class="MuiBox-root css-t7e7gw"]/div/ul/li')
        comfort_text = ' '.join([element.text for element in comfi_elements])
        comfort.append(comfort_text)
    except NoSuchElementException:
        comfort.append('NA')
```
